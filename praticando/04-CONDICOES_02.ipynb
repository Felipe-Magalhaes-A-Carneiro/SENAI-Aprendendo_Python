## 18 Exercícios para praticar condições em Python, pelo ChatGPT

### NÍVEL 1 — BÁSICO (Fundamentos do if)

Exercício 1 — Número positivo ou negativo

Leia um número inteiro e informe se ele é positivo, negativo ou zero.

```python
numero = int(input("Digite um número e diremos se ele é POSITIVO, NEGATIVO ou 0: "))

if (numero == 0):
    print(f"O número {numero} é 0 (ZERO)")
elif (numero > 0):
    print(f"O número {numero} é POSITIVO")
else:
    print(f"O número {numero} é NEGATIVO")
```

Exercício 2 — Par ou ímpar

Leia um número inteiro e informe se ele é par ou ímpar.

```python
numero = int(input("Digite um número e diremos se é PAR ou ÍMPAR: "))

if (numero % 2 == 0):
    print("O número {} é PAR".format(numero))
else:
    print("O número {} é ÍMPAR".format(numero))
```

Exercício 3 — Maior de idade

Leia a idade de uma pessoa e informe se ela é maior de idade (18 anos ou mais) ou menor de idade.

```python
idade = int(input("Digite a sua ídade e diremos se você é MAIOR ou MENOR de idade: "))

if (idade > 18):
    print(f"Você têm {idade} anos. É MAIOR de idade.")
else:
    print(f"Você têm {idade} anos. É MENOR de idade.")
```

Exercício 4 — Maior de dois números

Leia dois números inteiros e informe qual é o maior ou se são iguais.

```python
numero1 = int(input("Digite dois números e definiremos quais deles é o maior. Digite o primeiro número: "))
numero2 = int(input(f"Você digitou o número {numero1}. Agora digite o segundo número: "))

if (numero1 > numero2):
    print(f"Números digitados foram: {numero1} e {numero2}. O número maior é {numero1}")
else:
    print(f"Números digitados foram: {numero1} e {numero2}. O número maior é {numero2}")
```

### NÍVEL 2 — INTERMEDIÁRIO (Múltiplas condições)

Exercício 5 — Nota do aluno

Leia a nota de um aluno e informe:

    Aprovado (nota ≥ 7)

    Recuperação (5 ≤ nota < 7)

    Reprovado (nota < 5)

```python
nota = int(input("Digite a nota do aluno(a)"))

if (nota >= 7):
    print("O aluno(a) foi APROVADO.")
elif (nota == 5 or nota < 7):
    print("O aluno(a) está de RECUPERAÇÃO")
else:
    print("O aluno(a) foi REPROVADO")
```
