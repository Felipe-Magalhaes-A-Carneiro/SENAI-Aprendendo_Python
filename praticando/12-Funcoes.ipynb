```python
def adiciona_tres(numero):
   numero += 3 
   return numero

adiciona_tres(3)
```

```python
# Outra forma de fazê-lo com o mesmo resultado. Aqui havendo uma explicação:

def adiciona_4(numero_entrada):
    numero_saida = numero_entrada + 4
    return numero_saida

adiciona_4(20)
```

```python
# Chamando uma função dentro de uma variável.

novo_numero = adiciona_tres(10)

print(novo_numero)
```

```python
# Reaplicando a ideia anterior com uma forma mais complexa:

def recebe_pagamento(horas_trabalho):
    pagamento_dia = horas_trabalho * 15
    pagamento_com_taxa = pagamento_dia * (1 - .12)
    return pagamento_com_taxa

recebe_pagamento(40)
```

```python
def get_expected_cost(beds, baths):
    value = 80000
    value += beds + baths
    return value
    
get_expected_cost(30000,10000)
```
